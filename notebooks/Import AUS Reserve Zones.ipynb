```python
import os
from django.contrib.gis.gdal import DataSource

auszones = os.path.abspath('/Users/russmo/Code/wdpa/AustralianMarineParksEstate/AustralianMarineParks_merged.gpkg')
```

```python
ds = DataSource(auszones)
print(ds)
print(len(ds))
```

```python
for lyr in ds:
    print (lyr, lyr.geom_type, len(lyr), lyr.srs,)
    f_types = [fld.__name__ for fld in lyr.field_types]
```

```python
from mpa.models import Mpa

parklyr = ds['AustralianMarineParks_merged']
allfields = parklyr.fields.copy()
print(allfields)
for feat in parklyr:
    newgeom = feat.geom.clone()
    newgeom.transform(4326)
    m, created = Mpa.objects.get_or_create(other_ids='AUS:%s' % feat['POLYGONID'].value)
    m.name = '%s - %s' % (feat['RESNAME'].value, feat['ZONENAME'].value)
    m.orig_name = feat['RESNAME'].value
    m.sovereign = 'AUS'
    m.country = 'AUS'
    if feat['RESNAME'].value == 'Norfolk':
        m.country = 'NFK'
    if feat['NETNAME'].value == 'HIMI':
        m.country = 'HMD'
    m.iucn_category = feat['ZONEIUCN'].value
    m.rep_area = feat['AREA_KM2'].value
    m.geom = newgeom.geos
    m.save()
    m.refresh_from_db()
    m.categories.add('Australia CMR Network - %s' % feat['NETNAME'].value)
    print(m)
```
