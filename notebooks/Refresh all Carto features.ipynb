```python
from mpa.models import Mpa
from mpa.views import mpas_all_nogeom
from django.contrib.gis.db.models.functions import NumPoints
mpas = mpas_all_nogeom.annotate(numpoints=NumPoints('geom')).filter(numpoints__gte=50000)
mpas50000 = {}
for m in mpas:
    try:
        numgeom = m.numpoints
        if numgeom >= 50000:
            mpas50000[m.mpa_id] = numgeom
            # print(numgeom, m.name, m.designation)
    except:
        raise
bigmpas = {k:v for k,v in mpas50000.items() if v >= 500000}
len(bigmpas)
```
