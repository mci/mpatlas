```python
import os
from django.contrib.gis.gdal import DataSource
from importlib import reload

wdpa_201810 = os.path.abspath('/Users/russmo/Code/wdpa/WDPA_Oct2018_Public/WDPA_Oct2018_Public.gdb')
```

```python
ds = DataSource(wdpa_201810)
print(ds)
print(len(ds))
```

```python
def identify_layers(source=wdpa_201810):
    ds = DataSource(source)
    layers = {'point': None, 'poly': None, 'source': None}
    for lyrid in range(0, len(ds)):
        for key in layers.keys():
            if ds[lyrid].name.find(key) > -1:
                layers[key] = lyrid
                break
    return(layers)
```

```python
wdpa_point = ds[identify_layers()['point']]
wdpa_poly = ds[identify_layers()['poly']]
wdpa_source = ds[identify_layers()['source']]
for lyr in (wdpa_source, wdpa_point, wdpa_poly):
    print (lyr, lyr.geom_type, len(lyr), lyr.srs,)
    f_types = [fld.__name__ for fld in lyr.field_types]
    print (list(zip(lyr.fields, f_types, lyr.field_widths)))
```
