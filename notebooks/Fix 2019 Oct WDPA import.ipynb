```python
can_pids = Mpa.objects.filter(country='CAN', wdpa_pid__icontains='_')
can_pids.count()
```

```python
can_wids = can_pids.values_list('wdpa_id', flat=True).distinct('wdpa_id')
can_wids = list(set(can_wids))
len(can_wids)
```

```python
can_unimp = Mpa.objects.filter(wdpa_id__in=can_wids, implemented=False)
can_unimp.count()
```

```python
for c in can_unimp:
    c_zones = Mpa.objects.filter(wdpa_id=c.wdpa_id)
    for cz in c_zones:
        print(cz.wdpa_pid, cz.implemented, cz.name, cz.designation)
```
